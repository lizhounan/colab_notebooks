```
import numpy as np
import cvxpy as cvx
# data for censored fitting problem.
np.random.seed(15)
m = 20;  # dimension of x's
p = 25;  # number of non-censored data points
n = 100; # total number of points
w_true = np.random.randn(m,1)
X = np.random.randn(m,n)
y = np.dot(np.transpose(X),w_true) + 0.1*(np.sqrt(m))*np.random.randn(n,1)# Reorder measurements, then censor
sort_ind = np.argsort(y.T)
y = np.sort(y.T)
y = y.T
X = X[:, sort_ind.T]
D = (y[p-1]+y[p])/2.0
y = y[range(p)]
```

```
X = np.reshape(X, (20, 100))
y = np.reshape(y, (25, ))
w_true.reshape((20, ))
```

```
X_uncensored = X[:, :p]
X_censored = X[:, p:]
```

```
X_uncensored.shape
```

```

w = cvx.Variable(m)
# Construct the problem.
objective = cvx.Minimize(cvx.sum_squares(X_uncensored.T*w - y))
constraints = [X_censored.T*w >= D[0]]
prob = cvx.Problem(objective, constraints)

# The optimal objective value is returned by `prob.solve()`.
result = prob.solve()
# The optimal value for x is stored in `x.value`.
print(w.value)
```

```
w1 = w.reshape((20,1))
w2 = w_true.reshape((20,1))
error1 = np.dot((w1 - w2).T , (w1 - w2)) / np.dot(w1.T, w1)
print(error1)
```
