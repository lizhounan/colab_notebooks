```
from transformers import AlbertTokenizer, AlbertModel, AlbertForPreTraining, AutoConfig, AutoModelForSequenceClassification
import torch
import numpy as np
```

```
tokenizer = AlbertTokenizer.from_pretrained('albert-base-v2')
```

```
model = AlbertForPreTraining.from_pretrained('albert-base-v2', return_dict=True)
```

```
input_ids = torch.tensor(tokenizer.encode('Hello Huggingface Albert! This is my demo notebook', add_special_tokens=True)).unsqueeze(0)
```

```
import dataclasses
from dataclasses import dataclass, field
from typing import Dict, Optional
```

```
@dataclass
class ModelArguments:
  model_name_or_path: str = field(metadata={'help': 'path to pretrained model ot model identifier from huggingface.co/models'})
  config_name: Optional[str] = field(default=None, metadata={'help':'Pretrained config name or path if not the same as model_name'})
  tokenizer_name: Optional[str] = field(default=None, metadata={'help':'Pretrained tokenizer name or path if not the same as model_name'})
  cache_dir: Optional[str] = field(default=None, metadata={'help':'where you want to store the pretrained model downloaded from s3'})
```

```
model_args = ModelArguments(model_name_or_path='albert-base-v2')
```
